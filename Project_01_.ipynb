#***STEP 3 — Exploratory Data Analysis (EDA)***

### Question 1:
Which countries have the highest happiness scores?

```python
top_happy = df[['Country', 'Happiness Score']].sort_values(by='Happiness Score', ascending=False).head(10)
top_happy
```

### Observation

The analysis shows that countries such as Finland, Denmark, and Switzerland have the highest happiness scores. Most top-ranked countries belong to Europe.

##Question 2
Which countries have the lowest happiness scores?

```python
lowest_happy = df[['Country', 'Happiness Score']].sort_values(by='Happiness Score').head(10)
lowest_happy
```

### Observation

Countries with the lowest happiness scores are mainly affected by economic instability, low social support, and poor living conditions.

##Question 3
What is the average happiness score?

```python
average_score = df['Happiness Score'].mean()
print('Average Happiness Score:', average_score)
```

### Observation

The calculated average happiness score provides an overall understanding of global happiness levels across all countries in the dataset.

##Question 4
Which factor has the strongest correlation with happiness score?

```python
correlation = df.corr(numeric_only=True)
correlation['Happiness Score'].sort_values(ascending=False)
```

### Observation

The correlation analysis shows which factors are most strongly associated with happiness scores. Factors like GDP, social support, and healthy life expectancy generally show strong positive relationships.

##Question 5
Which regions are happiest on average?

```python
region_avg = df.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)
region_avg
```

### Observation

Some regions consistently show higher average happiness scores due to better economic conditions, healthcare, and social support systems.

#***STEP 4 — Visualizations***

##Chart 1 — Bar Chart
Top 10 Happiest Countries

```python
plt.figure(figsize=(10,6))
plt.bar(top_happy['Country'], top_happy['Happiness Score'])
plt.xticks(rotation=45)
plt.title('Top 10 Happiest Countries')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.show()
```

##Chart 2 — Line Chart
Happiness Scores Distribution

```python
plt.figure(figsize=(10,6))
plt.plot(df['Happiness Score'])
plt.title('Happiness Score Distribution')
plt.xlabel('Index')
plt.ylabel('Happiness Score')
plt.show()
```

##Chart 3 — Histogram
Distribution of Happiness Scores

```python
plt.figure(figsize=(10,6))
plt.hist(df['Happiness Score'], bins=20)
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

##Chart 4 — Scatter Plot
GDP vs Happiness Score

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'])
plt.title('GDP vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

##Chart 5 — Pie Chart
Top 5 Happiest Countries

```python
plt.figure(figsize=(8,8))

plt.pie(
    top_happy['Happiness Score'].head(5),
    labels=top_happy['Country'].head(5),
    autopct='%1.1f%%'
)

plt.title('Top 5 Happiest Countries')

plt.show()
```

##Chart 6 — Heatmap
Correlation Heatmap

```python
correlation = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```
